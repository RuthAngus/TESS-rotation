```python
import lightkurve as lk

starname = "TIC 10863087"
lcf = lk.search_lightcurvefile(starname).download()
```

```python
lc = lcf.PDCSAP_FLUX
lc.scatter(alpha=.5, s=.5);
```

```python
import pandas as pd
df = pd.DataFrame(dict({"time": lc.time, "flux": lc.flux}))
df.to_csv("test.csv")
```
